### changing the column name to readable dates

```python
def new_col(col):
    if '_N' in col:
        col= col[:-2]
        dt=pd.to_datetime(col)
        return dt.strftime('%d %b %Y')
    return col

df.columns=[new_col(col) for col in df.columns]
df.head()
```

### dealing with the outliers and the nan values

```python
def remove_outliers(df,col):
    classes=df['class'].unique()

    for class_name in classes:
        class_data=df[df['class']==class_name][col]
        Q1 = class_data.quantile(0.25)
        Q3 = class_data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        median = class_data.median()

        outlier_condition=((df['class']==class_name) & (df[col]> upper_bound)) | ((df['class']==class_name) & (df[col]<lower_bound))
        df.loc[outlier_condition,col]=median
    return df
# replacing the outliers with the median and also the nan values with the median

def remove_nan(df,col):
    classes=df['class'].unique()

    for class_name in classes:
        class_data=df[df['class']==class_name][col]
      
        median = class_data.median()

        nan_condition=((df['class']==class_name) & (df[col].isna())) 
        df.loc[nan_condition,col]=median
    return df
```

```python
ndvi_columns = df.columns[3:]  

for col in ndvi_columns:
    df = remove_outliers(df,col)
    df = remove_nan(df,col)

sns.heatmap(df.isna(),cmap='viridis',cbar=False)
#recheck for the nan values
```

###Training the model 

```python
df = df.drop(columns=['Unnamed: 0', 'ID'])
y = df['class']
X = df.drop(columns=['class'])


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)# ysing the scaler for the better classification

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print(classification_report(y_test, y_pred))

```

```python
train_pred = model.predict(X_train)
train_acc = accuracy_score(y_train, train_pred)
print("Train Accuracy:", train_acc)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```
